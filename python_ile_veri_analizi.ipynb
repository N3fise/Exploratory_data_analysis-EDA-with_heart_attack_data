```python
# 52. Veri seti ile ilgili olarak Scatter Plot grafiği çizer.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("veri.csv")  # Örnek CSV dosyası
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='chol', data=df)
plt.title('Yaş ve Kolesterol Scatter Plot')
plt.show()

```

```python
# 53. Histogram grafiği çizer.
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Yaş Histogramı')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.show()
```

```python
# 54. Bar chart grafiği çizer.
df['sex'].value_counts().plot(kind='bar', color='coral')
plt.title('Cinsiyet Dağılımı')
plt.xlabel('Cinsiyet')
plt.ylabel('Adet')
plt.show()
```

```python
# 55. Pie chart grafiği çizer.
df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Cinsiyet Dağılımı (Pasta Grafiği)')
plt.ylabel('')  # Y ekseni yazısını gizle
plt.show()
```

```python
# 56. Korelasyon matrisine göre heatmap çizer.
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Korelasyon Isı Haritası')
plt.show()
```

```python
# 57. Seaborn kütüphanesi ile stripplot çizer.
plt.figure(figsize=(8, 5))
sns.stripplot(x='sex', y='chol', data=df, jitter=True)
plt.title('Cinsiyete Göre Kolesterol (Stripplot)')
plt.show()
```

```python
# 58. Seaborn kütüphanesi ile swarmplot çizer.
plt.figure(figsize=(8, 5))
sns.swarmplot(x='sex', y='chol', data=df)
plt.title('Cinsiyete Göre Kolesterol (Swarmplot)')
plt.show()
```
